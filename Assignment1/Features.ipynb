```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Data/dataset_mood_smartphone.csv',index_col=0)
data
```

```python
# split up time into multiple columns
data[['date','timeOfDay']] = data.time.str.split(" ",expand=True,)
data[['year','month','day']] = data.date.str.split("-",expand=True,)
```

```python
# add column for phase of the day

# Morning (5 AM to 12 PM)
# Afternoon (12 PM to 5 PM)
# Evening (5 PM to 10 PM)
# Night (10 PM to 5 AM)

# morning,afternoon,evening,night = 0,1,2,3

phaseOfDay = []
for index, row in data.iterrows():
    time = int(row.timeOfDay.split(':')[0])
    if time >= 5 and time < 12:
        phaseOfDay.append(0)
    elif time >= 12 and time < 17:
        phaseOfDay.append(1)
    elif time >= 17 and time < 22:
        phaseOfDay.append(2)
    else:
        phaseOfDay.append(3)

data['phaseOfDay'] = phaseOfDay
```
