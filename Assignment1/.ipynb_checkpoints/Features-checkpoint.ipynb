```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Data/dataset_mood_smartphone.csv',index_col=0)
data
```

```python
# split up time into multiple columns
data[['date','timeOfDay']] = data.time.str.split(" ",expand=True,)
data[['year','month','day']] = data.date.str.split("-",expand=True,)
```

```python
# create data structure that helps determine day of week

# 2014 dates
# February (sat 1 - fri 28)
# March (sat 1 - mon 31)
# April (tues 1 - wed 30)
# May (thurs 1 - sat 31)
# June (sun 1 - mon 30)

# mon,tues,wed,thurs,fri,sat,sun = 1,2,3,4,5,6,7

# key is month
# value is [weekday of first day, length of month]
calendar = {
    '02' : [6,28], 
    '03' : [6,31],
    '04' : [2,30], 
    '05' : [4,31], 
    '06' : [7,30], 
}

days = {}
for key,value in calendar.items():
    weekdays = []
    counter = value[0]
    for day in range(value[1]):
        mod = counter % 7
        if mod == 0:
            weekdays.append(7)
        else:
            weekdays.append(mod)
        counter += 1
    days[key] = weekdays

print("")
for k,v in days.items():
    print(k,v)
```

```python
# add column for day of week

weekday = []
for index, row in data.iterrows():
    weekday.append(days[row.month][int(row.day)-1])

data['weekday'] = weekday
data
```

```python
# add column for phase of the day

# Morning (5 AM to 12 PM)
# Afternoon (12 PM to 5 PM)
# Evening (5 PM to 10 PM)
# Night (10 PM to 5 AM)

# morning,afternoon,evening,night = 1,2,3,4

phaseOfDay = []
for index, row in data.iterrows():
    time = int(row.timeOfDay.split(':')[0])
    if time > 5 and time < 12:
        phaseOfDay.append(1)
    elif time > 12 and time < 17:
        phaseOfDay.append(2)
    elif time > 17 and time < 22:
        phaseOfDay.append(3)
    else:
        phaseOfDay.append(4)

data['phaseOfDay'] = phaseOfDay
data
```
